## Runge-Kutta Methods

### 2nd-order Runge-Kutta

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, t):
    return t + 2*x*t


def RK2(f, t, h):
    
    for i in range(0, len(x)-1):
        k1 = h*f(x[i], t[i])
        k2 = h*f(x[i] + h*k1/2, t[i] + h/2)
        x[i+1] = x[i] + k2
    
    return x

h = 0.01
t = np.arange(0, 2+h, h)
x = np.zeros_like(t)

plt.title("2nd-order Runge-Kutta Method")
plt.plot(t, RK2(f, t, h))
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

### 4th-order Runge-Kutta

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, t):
    return t + 2*x*t


def RK4(f, t, h):
    
    for i in range(0, len(x)-1):
        k1 = h*f(x[i], t[i])
        k2 = h*f(x[i] + h*k1/2, t[i] + h/2)
        k3 = h*f(x[i] + h*k2/2, t[i] + h/2)
        k4 = h*f(x[i] + k3, t[i] + h)
        x[i+1] = x[i] + (k1 + 2*k2 + 2*k3 + k4)/6
    
    return x

h = 0.01
t = np.arange(0, 2+h, h)
x = np.zeros_like(t)

plt.title("4th-order Runge-Kutta Method")
plt.plot(t, RK4(f, t, h))
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

### Analysis of Order of Accuracy

```python
# Exact Soluton
def exact(t):
    return 0.5*(np.exp(t**2)-1)

# Plot Solution's Error
plt.figure()
plt.title("Solution's Error")
plt.plot(t, np.abs(exact(t) - RK2(f, t, h)), label='RK2')
plt.plot(t, np.abs(exact(t) - RK4(f, t, h)), label='RK4')
plt.xlabel("t")
plt.ylabel("E")
plt.legend()
plt.show()
```

### Effect of Step Size

```python
import numpy as np
import matplotlib.pyplot as plt

# Exact Soluton
def exact(t):
    return 0.5*(np.exp(t**2)-1)

# Given Function
def f(x, t):
    return t + 2*x*t

# 4nd-order Runge-Kutta Method
def RK4(f, t):
    
    for i in range(0, len(x)-1):
        k1 = h*f(x[i], t[i])
        k2 = h*f(x[i] + h*k1/2, t[i] + h/2)
        k3 = h*f(x[i] + h*k2/2, t[i] + h/2)
        k4 = h*f(x[i] + k3, t[i] + h)
        x[i+1] = x[i] + (k1 + 2*k2 + 2*k3 + k4)/6
    
    return x

# Calculate Values and Plot
for h in [0.01, 0.05, 0.1]:
    t = np.arange(0, 2+h, h)
    x = np.zeros_like(t)
    if h == 0.01:
        t_h1 = t
        x_h1 = RK4(f, t)
        x_exact1 = exact(t)
    elif h == 0.05:
        t_h2 = t
        x_h2 = RK4(f, t)
        x_exact2 = exact(t)
    else:
        t_h3 = t
        x_h3 = RK4(f, t)
        x_exact3 = exact(t)

plt.figure()
plt.title("Error according to Step Size")
plt.plot(t_h1, np.abs(x_exact1 - x_h1), label="h=0.01")
plt.plot(t_h2, np.abs(x_exact2 - x_h2), label="h=0.05")
plt.plot(t_h3, np.abs(x_exact3 - x_h3), label="h=0.1")
plt.xlabel("t")
plt.ylabel("error")
plt.yscale('log')
plt.legend()
plt.show()

print("Error according to Step Size of h=0.01 at t=2")
print(f">> {np.abs(x_exact1[-1] - x_h1[-1])}")
print("Error according to Step Size of h=0.05 at t=2")
print(f">> {np.abs(x_exact2[-1] - x_h2[-1])}")
print("Error according to Step Size of h=0.1 at t=2")
print(f">> {np.abs(x_exact3[-1] - x_h3[-1])}")

plt.figure()
plt.title("Error according to Step Size at t=2")
plt.scatter(t_h1[-1], np.abs(x_exact1[-1] - x_h1[-1]), label="h=0.01")
plt.scatter(t_h2[-1], np.abs(x_exact2[-1] - x_h2[-1]), label="h=0.05")
plt.scatter(t_h3[-1], np.abs(x_exact3[-1] - x_h3[-1]), label="h=0.1")
plt.xlabel("t")
plt.ylabel("error")
plt.yscale('log')
plt.legend()
plt.show()
```

## Baseball Dynamics

```python
import numpy as np
import matplotlib.pyplot as plt

''' 1. Parameters '''
B = 4.1E-4       # Magnus Force Dimensionless Coefficient
omega = 1800.0   # Rotation Rate [rpm]
g = 9.81         # Gravity Acceleration [m/s^2]
dt = 0.01        # Time Step [s]
h = 1.7          # Height of Initial Condition [m]
v0 = 40.0        # Initial Speed [m/s]
theta = 1.0      # Upward Angle [degree]
phi = 225.0      # Rotation Angle about z-axis [degree]
time = 0.0       # Initial Time


''' 2. Equations '''
def drag(V):
    return 0.0039 + 0.0058/(1 + np.exp((V - 35)/5))

def accel(drag, vx, vy, vz):
    F = drag(V)
    ax = -F*V*vx + B*omega*(vz*np.sin(np.deg2rad(phi))-vy*np.cos(np.deg2rad(phi)))
    ay = -F*V*vy + B*omega*vx*np.cos(np.deg2rad(phi))
    az = -g - F*V*vz - B*omega*vx*np.sin(np.deg2rad(phi))
    return ax, ay, az


''' 3. Initial Conditions '''
x, y, z = 0.0, 0.0, h
vx, vy, vz = v0*np.cos(np.deg2rad(theta)), 0.0, v0*np.sin(np.deg2rad(theta))

list_x, list_y, list_z = [x], [y], [z]


''' 4. Time Integration using RK4 Scheme '''
while x <= 18.39:
    print("========================================================")
    print(f"                  Time step : {time:.2f}s")
    print("========================================================")
    print(f" >> x  : {x:.2f}m   | y  : {y:.2f}m   | z  : {z:.2f}m")
    print("--------------------------------------------------------")
    print(f" >> vx : {vx:.2f}m/s | vy : {vy:.2f}m/s | vz : {vz:.2f}m/s")
    print("--------------------------------------------------------")
    print(f" >> V : {V:.2f}m/s")
    print("========================================================")
    
    V = np.sqrt(vx**2 + vy**2 + vz**2)
    ax, ay, az = accel(drag, vx, vy, vz)
        
    # 4th-order Runge-Kutta Method
    xk1 = accel(drag, vx, vy, vz)[0]
    xk2 = accel(drag, (vx + dt*xk1/2), vy, vz)[0]
    xk3 = accel(drag, (vx + dt*xk2/2), vy, vz)[0]
    xk4 = accel(drag, (vx + xk3), vy, vz)[0]
    vx += dt*(xk1 + 2*xk2 + 2*xk3 + xk4)/6
    
    yk1 = accel(drag, vx, vy, vz)[1]
    yk2 = accel(drag, vx, (vy + dt*yk1/2), vz)[1]
    yk3 = accel(drag, vx, (vy + dt*yk2/2), vz)[1]
    yk4 = accel(drag, vx, (vy + yk3), vz)[1]
    vy += dt*(yk1 + 2*yk2 + 2*yk3 + yk4)/6
    
    zk1 = accel(drag, vx, vy, vz)[2]
    zk2 = accel(drag, vx, vy, (vz + dt*zk1/2))[2]
    zk3 = accel(drag, vx, vy, (vz + dt*zk2/2))[2]
    zk4 = accel(drag, vx, vy, (vz + zk3))[2]
    vz += dt*(zk1 + 2*zk2 + 2*zk3 + zk4)/6
    
    # Calculate Position
    x += vx*dt
    y += vy*dt
    z += vz*dt
    
    # Save Position
    list_x.append(x)
    list_y.append(y)
    list_z.append(z)
    
    time += dt

print("====================== End Solver ======================")

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(list_x, list_y, list_z)
ax.plot(list_x, list_y, z-6.4, 'r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.text(0, 0, h, 'Start')
ax.set_xlim(0, 19)
ax.set_ylim(-5, 5)
ax.set_zlim(-3, 3)
plt.show()
```
